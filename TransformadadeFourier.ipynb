```python
import numpy as np
import matplotlib.pyplot as plt

################################################
# Questão 1 - Letra A
################################################
def sinal(t):
    if -0.5 <= t <= 0.5:
        valor = 1 - np.abs(t)
    else:
        valor = 0
    return valor

# Vetor de tempo
t = np.linspace(-2, 2, 1000)

# Calcula os valores da função para cada ponto no vetor de tempo
sinal = np.array([sinal(ti) for ti in t])

# Plotar o sinal
plt.plot(t, sinal, 'b')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.title('Sinal 1 - |t|')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definição da forma de onda desejada no item 1a)
def onda(t):
    if -0.5 <= t <= 0.5:
        valor = 1 - np.abs(t)
    else:
        valor = 0.0
    return valor

def MS7P1(x):
    y = np.ones_like(x)
    i = np.where(x != 0)
    y[i] = np.sin(x[i]) / x[i]
    return y

def X(w, tau):
    return ((tau*(MS7P1(w*tau/2) / 2)) +  (tau/2)*(MS7P1(w*tau/4)*2 /2 ))

omega = np.linspace(-8*np.pi, 8*np.pi, 1000)

plt.plot(omega, X(omega, 1), 'k-', label='tau=-1')

plt.grid(True)
plt.xlabel('\u03C9')
plt.ylabel('X(\u03C9)')
plt.legend()
plt.show()

x = np.linspace(-2, 2, 1000)
onda_vectorizada = np.vectorize(onda, otypes=[float])
y = onda_vectorizada(x)

plt.plot(x, y, label=("Original"))
plt.xlabel('t')    # omega
plt.ylabel('X(t)') # X(omega)
plt.legend()
plt.show()
```

```python
import numpy as np

def largura_faixa(beta, a):
    W = np.tan(beta*np.pi/2) * a
    return W

beta1 = 0.95
a1 = 1
W1 = largura_faixa(beta1, a1)
print("W1:", W1)

beta2 = 0.90
a2 = 2
W2 = largura_faixa(beta2, a2)
print("W2:", W2)

beta3 = 0.5
a3 = 3
W3 = largura_faixa(beta3, a3)
print("W3:", W3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rect(x):
    return np.where(np.abs(x) < 0.5, 1, 0)

# |t| * rect((t-1)/3)
def signal(t):
    return np.abs(t) * rect((t-1)/3)

# Define the time range
t = np.linspace(-5, 5, 1000)
s = signal(t)

# Plot the signal
plt.plot(t, s)
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.title('|t| * rect((t-1)/3)')
plt.grid(True)
plt.show()
```
