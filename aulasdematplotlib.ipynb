```
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2*np.pi, 1000)
y = np.sin(4*t)

plt.plot(t, y)
plt.title("Gr√°fico do cosseno")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.show() #gera a figura
```

```
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2*np.pi, 1000)
y = np.cos(4*t)
y1 = np.sin(4*t)

plt.figure("Cosseno")
plt.plot(t, y)
plt.title("Gr√°fico do cosseno")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.show() #gera a figura

plt.figure(2)
plt.plot(t, y1)
plt.title("Gr√°fico do seno")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.show() #gera a figura
```
