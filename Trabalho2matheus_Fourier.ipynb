```python
import numpy as np
import matplotlib.pyplot as plt

def cutted_triangle_wave(x, period=np.pi, height=1, ratio=3):
    x = x % period  # make x value always between 0 and period
    result = []

    for t in x:
        if t > period / ratio:
            result.append(0)  # zeroed part of the function
        else:
            # weighted ratio a*t line slope
            result.append(t * (height * ratio / period))

    # return single value, or vector with size of x
    return np.asarray(result)

# Criando um vetor de valores de x
x = np.linspace(0, 2*np.pi, 1000)

# Definindo os parâmetros da forma de onda
period = np.pi
height = 1
ratio = 3

# Calculando a forma de onda triangular cortada
wave = cutted_triangle_wave(x, period, height, ratio)

# Plotando a forma de onda triangular cortada
plt.plot(x, wave)
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.title('Forma de Onda Triangular Cortada')
plt.grid(True)
plt.show()
```
