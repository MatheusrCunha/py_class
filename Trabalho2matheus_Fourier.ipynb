```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, sqrt
from sympy import And

T = pi  #Período
w = (2.0 * pi) / T  #Frequência Ângular
intervalo = np.arange(-5, 5, 0.001) #cria um array valores no intervalo de -5 a 5, com um passo de 0.001
res = np.zeros_like(intervalo, dtype=complex) # zeros complexos.
# Definindo os parâmetros da forma de onda
period = np.pi
height = 1
ratio = 3

coef = input("Quantos Coeficientes deseja? ")  #quantos coeficientes você deseja
n = int(coef) #Coeficiente será inteiro

##################################################
# Definindo os parâmetros da forma de onda do sinal P6.1-6 (a)
period = pi
height = 1
ratio = 3

# Criando um vetor de valores de x
x = np.arange(-5, 5, 0.001)

# Criando um array para armazenar os valores da forma de onda triangular cortada
wave = np.zeros_like(x)

# Preenchendo o array com os valores da forma de onda triangular cortada
for i, t in enumerate(x):
    t = t % period
    if t > period / ratio:
        wave[i] = 0
    else:
        wave[i] = t * (height * ratio / period)
########################################################
# Definindo os parâmetros da forma de onda do sinal P6.1-6 (b)
########################################################
i = 0
for t in intervalo:
    valor = 0.0
    for k in range(-n, n+1):
        if k != 0:
            valor += (((1j*k*w - np.exp(1j*k*w) + 1) * (np.exp(1j*k*w*t-1j*k*w))) / (pi*k**2*w**2))
        else:
            valor += 0.15915
    res[i] = valor
    i += 1

# Plotando a forma de onda triangular cortada
plt.plot(x, wave, label='Forma de Onda Triangular')
plt.plot(intervalo, res.real, label='Sinal P6.1-6 (a) - Lathi')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.title('Forma de Onda Triangular e Sinal P6.1-6 (a) - Lathi')
plt.legend()
plt.grid(True)
plt.show()

#########################################################
i=0
for t in intervalo:
    valor = 0.0
    for k in range(-n, n+1):
        if k != 0:
            valor += (((1j*k*w - 2*np.exp(1j*k*w) + np.exp(2*1j*k*w) * (1 - 1j*k*w) + 1) * np.exp(-1j*k*(t+1)*w)) / (pi*k**2*w**2))
        else:
            valor += 0.31831

    res[i] = valor
    i += 1

plt.ylabel('Amplitude')
plt.title('Amplitude da Função Onda(t)')
plt.ylabel('Sinal P6.1-6 (b) - Lathi')
plt.plot(intervalo, res.real)
plt.legend()
plt.grid(True)
plt.show()
########################################################
n = 10 #Coeficientes para o espectro de fourier
# Vetor de valores de k
k_values = np.arange(-n, n+1)

# Vetores para armazenar as magnitudes e fases dos coeficientes
magnitudes = []
fases = []
magnitudes_2 = []
fases_2 = []

for k in k_values:
    if k != 0:
        coef = (((np.exp(-1j*k*w)) * (1j*w*k - np.exp(1j*k*w) + 1)) / (pi*k**2*w**2))
        magnitude = np.abs(coef)
        magnitudes.append(magnitude)
    else:
        magnitudes.append(0.15915)  # Coeficiente a0 para k = 0

# Plotar o espectro de magnitude
plt.stem(k_values, magnitudes, basefmt='b-')
plt.xlabel('Harmônicos')
plt.ylabel('Magnitude')
plt.title('Espectro de Magnitude -  P6.1-6 (a) - Lathi')
plt.grid(True)
plt.show()

for k in k_values:
    if k != 0.0:
        coef = (((np.exp(-1j*k*w)) * (1j*w*k - np.exp(1j*k*w) + 1)) / (pi*k**2*w**2))
        fase = np.angle(coef)
        fases.append(fase)
    else:
        fases.append(0.15915)  # Coeficiente a0 para k = 0

# Plotar o espectro de fase
plt.stem(k_values, fases, basefmt='b-')
plt.xlabel('Harmônicos')
plt.ylabel('Fase (rad)')
plt.title('Espectro de Fase -  P6.1-6 (a) - Lathi')
plt.grid(True)
plt.show()

#################################################################
for k in k_values:
    if k != 0:
        coef = ((np.exp(-1j*k*w)) * (1j*k*w - 2*np.exp(1j*k*w) + np.exp(2*1j*k*w)*(1 - 1j*k*w) + 1) / (pi*k**2*w**2))
        magnitude = np.abs(coef)
        magnitudes_2.append(magnitude)
    else:
        magnitudes_2.append(0.31831)  # Coeficiente a0 para k = 0

# Plotar o espectro de magnitude
plt.stem(k_values, magnitudes_2, basefmt='b-')
plt.xlabel('Harmônicos')
plt.ylabel('Magnitude')
plt.title('Espectro de Magnitude -  P6.1-6 (b) - Lathi')
plt.grid(True)
plt.show()

for k in k_values:
    if k != 0.0:
        coef = ((np.exp(-1j*k*w)) * (1j*k*w - 2*np.exp(1j*k*w) + np.exp(2*1j*k*w)*(1 - 1j*k*w) + 1) / (pi*k**2*w**2))
        fase = np.angle(coef)
        fases_2.append(fase)
    else:
        fases_2.append(0.31831)  # Coeficiente a0 para k = 0

# Plotar o espectro de fase
plt.stem(k_values, fases_2, basefmt='b-')
plt.xlabel('Harmônicos')
plt.ylabel('Fase (rad)')
plt.title('Espectro de Fase -  P6.1-6 (b) - Lathi')
plt.grid(True)
plt.show()




```
