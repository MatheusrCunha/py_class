```python
import numpy as np # biblioteca de processamento numérico
from scipy import signal # biblioteca de processamento de sinais
import matplotlib.pyplot as plt # biblioteca para visualização de dados
```

Impulso discreto $δ[n]:$

O sinal abaixo é um trecho de um sinal impulso unitário. O sinal é definido como zeros para todas as amostras negativas e positivas, se estendendo assim até $-∞$ e $+\infty$, respectivamente. O valor referente à amostra 0 é 1.

```python
x = signal.unit_impulse(9,4) # O primeiro argumento é o comprimento do sinal. O segundo argumento define a posição do impulso. Temos que lembrar que o python conta elementos a partir do 0 (diferente do matlab)
n = np.linspace(-4,4,9) #np.linspace(início, fim , número de amostras)
plt.figure()
plt.grid()
plt.stem(n,x)
plt.title('trecho de função impulso')
```

Degrau unitário:
Como o degrau unitário tem infinitas amostras unitárias a partir do índice zero, rigorosamente não é possível representá-lo como um trecho de sinal de duração finita (exceto com "três pontinhos" após as amostras).
Por outro lado, podemos calcular a resposta ao degrau unitário através, por exemplo, da função *step*, obtida do submódulo *signal* da biblioteca *scipy*.

Se você quiser, pode criar um segmento do sinal, mas isso não substitui o uso do sinal conforme a definição.

```python
x = np.zeros(100) # cria um vetor de zeros com 100 elementos
x[50:100] = 1. # substitui por 1 os valores referentes aos índices a partir de 50
n = np.arange(-50,50,1) # cria um vetor amostral de -50 até 50 com o numpy.arange (início (fechado), final (aberto), passo)
plt.figure()
plt.grid()
plt.stem(x)
plt.title('Trecho de função degrau unitário')
```

```python
# Alternativamente:
x = np.ones(100)
x[0:49] = 0
plt.figure()
plt.grid()
plt.stem(n,x)
plt.title('Trecho de função degrau unitário')
```

Função exponencial x[n] = $a^{n}$, com $a$ real.

Se $\left|a\right|>1$, a magnitude do sinal cresce exponencialmente com n.

```python
n = np.arange(-5.,6,1)
a = 2
x = a**n
plt.figure()
plt.grid()
plt.stem(n,x)
plt.title('Trecho de função degrau')
```

Se $\left|a\right|<1$, a magnitude do sinal decai exponencialmente com n.

```python
a = 0.5
x = a**n
plt.figure()
plt.grid()
plt.stem(n,x)
plt.title('Trecho de função degrau')
```


Se $a>0$, o sinal se mantém positivo.

```python
a = 1.5
x = a**n
plt.figure()
plt.grid()
plt.stem(n,x)
plt.title('Trecho de função degrau')
```

Se $a$ é negativo, o sinal alterna entre valores positivos e negativos.

```python
a = -0.5
x = a**n
plt.figure()
plt.grid()
plt.stem(n,x)
plt.title('Trecho de função degrau')
```
