```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-1,4,1) # criação de índices amostrais auxiliares
print(n)
```

Criação de dois sinais discretos como vetores:

```python
x1 = np.array([-1,0,2,1,1]) # exemplo de vetor linha
x2 = np.array([0, -1, -1, 1, -2])
print(x1) # retorna x1
print(x2)
```

```python
plt.figure()
plt.stem(n,x1)
plt.title('x1')
plt.grid()
plt.figure()
plt.stem(n,x2)
plt.title('x2')
plt.grid()
```

# Soma:

```python
x3 = x1+x2
plt.figure()
plt.stem(n,x3)
plt.grid()
plt.title('x3')
```

# Multiplicação:

```python
x4 = x1*x2
plt.figure()
plt.stem(n,x4)
plt.grid()
plt.title('x4')
```

Composição de uma função com 3 elementos unitários, referentes aos índices amostrais 0, 1 e 2, usando uma degraus unitários:

```python
# Aluno: Ueslei Marian
# função x=u[n]-u[n-3] para o intervalo de -4 a 6.

def x(i):
  N=100
  u = np.zeros(N) # cria um vetor de zeros com N elementos
  u[50:N] = 1. # substitui por 1 os valores referentes aos índices a partir de 50
  n=50+i
  x=u[n]-u[n-3]
  return x

n = np.arange(-4,11,1) # cria um vetor amostral de -50 até 50 com o numpy.arange (início (fechado), final (aberto, que não faz parte da quantidade), passo)
plt.figure()
plt.grid()
plt.stem(n,x(n))
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Trecho de função degrau unitário')
```

## Convolução:


Exemplo:

Considere que o sinal $h[n]$, a seguir, representa a resposta ao impulso de um sistema linear invariante no tempo.

```python
h = np.array([.6, .5, .4, .3, .2, .1])
n_h = np.arange(0,6,1)
plt.stem(n_h,h)
plt.xlabel('$n$ [amostras]')
plt.ylabel('$h [n]$')
```

## Convolução com impulso unitário e convolução com impulso unitário com atraso



Criando índices amostrais entre 0 e 3:


```python
n = np.arange(0,4,1) # Lembrando, os limites da função arrange, do numpy, são definidos pelos dois primeiros argumentos, sendo aberta no segundo argumento (não o contém). O passo é dado pelo teceiro argumento.
x = np.zeros(len(n))
x[0] = 1.
plt.stem(n,x)
```

```python
y = np.convolve(x,h)
plt.stem(y)
```

Criando índices amostrais entre -3 e 3 e organizando dados auxiliares úteis:


```python
n = np.arange(-3,4,1)
x = np.zeros(len(n))
x[3] = 1.
plt.stem(n,x)
```

```python
y = np.convolve(x,h)
plt.stem(y)
```
